# 파일 읽기

```python
property_master_excel_path = "../input/건축물대장_kapt_병합.xlsx"
property_master_csv_path = "../input/건축물대장_kapt_병합.csv"

sapt_excel_path = "../input/서울시 공공주택 아파트 정보.xlsx"
sapt_csv_path = "../input/sapt.csv"
```

```python
master = pd.read_csv(property_master_csv_path, index = False)
print(f"master : {len(master)}")

sapt = pd.read_csv(sapt_csv_path, index = False)
print(f"sapt : {len(sapt)}")
```

# 병합

```python
merged = master.merge(
    sapt,
    left_on='단지코드',
    right_on='k-아파트코드',
    how='outer',
    indicator=True
)

matched = (
    merged.loc[merged['_merge'] == 'both']
    .drop(columns = ['_merge'])
    .copy()
)

unmatched_master = (
    merged.loc[merged['_merge'] == 'left_only']
    .drop(columns = ['_merge'])
    .copy()
)

unmatched_sapt = (
    merged.loc[merged['_merge'] == 'right_only']
    .drop(columns = ['_merge'])
    .copy()
)

print(f"merged : {len(merged)}")
print(f"matched : {len(matched)}")
print(f"unmatched_master = {len(unmatched_master)}")
print(f"unmatched_sapt : {len(unmatched_sapt)}")
```

```python
matched.to_excel("../output/건축물대장_kapt_sapt_병합됨.xlsx", index = False)
unmatched_master.to_excel("../output/건축물대장_kapt_병합실패.xlsx", index = False)
unmatched_sapt.to_excel("../output/sapt_병합실패.xlsx", index = False)
```
