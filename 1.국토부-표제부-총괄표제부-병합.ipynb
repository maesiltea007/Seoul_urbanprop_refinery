나머지는 국토부에서 공개한 DB data type 과 1:1 대응된다.

```python
표제부_path = f'./input/일반{년도}.txt'
총괄표제부_path = f'./input/총괄{년도}.txt'
```

## 국토부 건축물대장 - 표제부

```python
# First read all columns as string
표제부 = pd.read_csv(
    표제부_path, 
    dtype='string',  # Read everything as string first
    encoding='UTF-8',     
    sep='|', 
    na_values=['', 'NULL', 'null', 'NaN', 'nan'], 
    names=list(표제부_dtypes.keys()),
    header=None,
    index_col=False,
    skiprows=0
)
```

# 총괄표제부/표제부 합치기

총괄표제부는 법적으로는 필지 내 건물이 2개 이상 있을 경우 작성되어야 하나, 사람이 직접 작성하는 것이기 때문에 누락된 경우가 많다. 따라서, 총괄표제부 - 표제부 규칙을 일관되게 적용하기 위해 총괄표제부가 누락된 필지의 경우 임의로 생성한 총괄표제부 row 를 생성했다.

```python
# 1. 모든 컬럼 포함해서 합치기
combined_df = pd.concat([총괄표제부, 표제부], axis=0, ignore_index=True)

# 2. 총괄표제부 여부 표시 ('대장종류코드' 가 1 인 경우)
combined_df['is_총괄'] = combined_df['대장종류코드'].eq('1')

# 3. 총괄표제부가 없는 그룹 찾기
group_stats = combined_df.groupby('대지위치').agg({
    'is_총괄': 'sum',
    '관리건축물대장PK': 'count'
}).reset_index()

# 총괄표제부가 없고 표제부가 2개 이상인 그룹 찾기
needs_총괄 = group_stats[
    (group_stats['is_총괄'] == 0) & 
    (group_stats['관리건축물대장PK'] > 1)
]['대지위치'].tolist()  # Series 대신 list로 변환

# 4. 새로운 총괄표제부 row 생성
if needs_총괄:
    도로명_mapping = combined_df[
        combined_df['대지위치'].isin(needs_총괄)
    ].groupby('대지위치')['도로명대지위치'].first().to_dict()
    
    new_총괄_rows = pd.DataFrame([
        {
            '관리건축물대장PK': f"00000-{hash(대지위치) % 1000000:06d}",
            '대지위치': 대지위치,
            '도로명대지위치': 도로명_mapping.get(대지위치, ''),
            'is_총괄': True
        }
        for 대지위치 in needs_총괄
    ])

    print(f'새롭게 생성된 총괄표제부 행 수: {len(new_총괄_rows)}')
    
    combined_df = pd.concat([combined_df, new_총괄_rows], ignore_index=True)

# 대지위치별 건물 수 계산
건물수_by_대지위치 = combined_df.groupby('대지위치').size()

# '그룹 종류' 컬럼 생성
def determine_그룹종류(row):
    if row['is_총괄'] == True or row['대장종류코드'] == '1':
        return 0
    elif 건물수_by_대지위치[row['대지위치']] > 1:
        return 1
    else:
        return 2

combined_df['그룹 종류'] = combined_df.apply(determine_그룹종류, axis=1)

desired_columns = ['관리건축물대장PK', '그룹 종류'] + [col for col in combined_df.columns if col not in ['관리건축물대장PK', '그룹 종류']]
combined_df = combined_df[desired_columns]

# 5. 최종 정렬
combined_df = combined_df.sort_values(
    ['대지위치', '그룹 종류'], 
    ascending=[True, True]
)

# 6. 불필요한 컬럼 제거
combined_df = combined_df.drop('is_총괄', axis=1)
```

```python
# 7. 병합_dtypes에 맞게 컬럼 구성 및 데이터 타입 변환 (순서 중요)
combined_df = combined_df.reindex(columns=list(병합_dtypes.keys()))

# 데이터 타입 변환
for col, dtype in 병합_dtypes.items():
    if dtype in ['Int64', 'Float64']:
        combined_df[col] = pd.to_numeric(combined_df[col], errors='coerce').astype(dtype)
    else:
        combined_df[col] = combined_df[col].astype(dtype)   
```

# 총괄표제부에서 누락된 연면적/대지면적 계산하기

총괄표제부는 같은 대지위치에 있는 표제부들을 포함한 대지면적과, 이들의 연면적을 합산한 연면적을 가지고 있어야 한다. 하지만 누락된 경우가 많으므로 직접 채워넣었다.

```python
# 만일 총괄표제부 내에 대지면적이 없거나 0일 경우, 같은 대지위치에 있는 표제부('그룹 종류' == 1)의 대지면적 중 가장 큰 값을 복사한다.
max_대지면적 = combined_df[combined_df['그룹 종류'] == 1].groupby('대지위치')['대지면적(㎡)'].max()

mask = (combined_df['그룹 종류'] == 0) & (combined_df['대지면적(㎡)'].isna() | (combined_df['대지면적(㎡)'] == 0))
combined_df.loc[mask, '대지면적(㎡)'] = combined_df.loc[mask, '대지위치'].map(max_대지면적)
```

```python
# 만일 총괄표제부 내에 연면적이 없거나 0일 경우, 같은 대지위치에 있는 표제부('그룹 종류' == 1)의 연면적을 모두 합산해 대입한다.
sum_연면적 = combined_df[combined_df['그룹 종류'] == 1].groupby('대지위치')['연면적(㎡)'].sum()

mask = (combined_df['그룹 종류'] == 0) & (combined_df['연면적(㎡)'].isna() | (combined_df['연면적(㎡)'] == 0))
combined_df.loc[mask, '연면적(㎡)'] = combined_df.loc[mask, '대지위치'].map(sum_연면적)
```
