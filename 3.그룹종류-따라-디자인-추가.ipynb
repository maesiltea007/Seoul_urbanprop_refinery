# 액셀 파일의 경우 '그룹 구분' 디자인 추가

'그룹 구분'이 0번일 경우 진한 초록색, 1번일 경우 연두색, 2번일 경우 파란색으로 '그룹 구분' 컬럼의 셀 구성

```python
def color_groups(val):
    colors = {
        0: 'background-color: #00B050',  # 진한 초록색
        1: 'background-color: #92D050',  # 연두색
        2: 'background-color: #0070C0'   # 파란색
    }
    return colors.get(val, '')

# 스타일 적용하여 Excel 저장
styled_df = df.style.apply(lambda x: [color_groups(v) if x.name == '그룹 종류' else '' for v in x])
```

```python
#테스트용 엑셀파일 생성
styled_df[:10000].to_excel(f'./output/테스트용-표제부-총괄표제부-병합및정제-디자인추가-{년도}.xlsx', index=False)
```
